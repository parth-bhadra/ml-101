## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y=[11,22,33,44]

#2. Setup plt
fig,ax = plt.subplots(figsize=(10,10)) # width, height

# 3. Plot data
ax.plot(x, y)

# 4. Customise plot
ax.set(title ="Simple plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show 
fig.savefig("images/sample-plot.png")
```

## Making figures with NumPy arrays
We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
import numpy as np
```

```python
# create some data
x= np.linspace(0,10,100)
x
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# use same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
nut_butter_prices = {
    "Almond butter": 10,
    "Peanut butter": 8,
    "Cashew butter": 12
}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values(), 0.5)
ax.set(title="Price chart",
      xlabel="Item",
      ylabel="$");
```

```python
# making a horizontal chart
fig, ax =plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
x = np.random.randn(1000)
fig, ax =plt.subplots()
ax.hist(x);
```
