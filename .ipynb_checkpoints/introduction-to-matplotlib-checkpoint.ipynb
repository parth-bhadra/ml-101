# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plt
fig, ax = plt.subplots(figsize=(10,10)) # width, height

# 3. plot data
ax.plot(x, y)

# 4. Customise plot
ax.set(title = "Simple plot",
      xlabel = "x-axis",
      ylabel = "y-axis")
# 5. Save and show
fig.savefig("images/sample-plot.png")
```

```python
# make a plot from a disctionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",
      ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# make some data and plot a histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# subplot option 1
fig, ((ax00,ax01), (ax10,ax11)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#plot to each different axis
ax00.plot(x, x/2);
ax01.scatter(np.random.random(10), np.random.random(10));
ax10.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax11.hist(np.random.randn(1000));
```

```python
# subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
# plot to each different index

ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib oo mehtod)
* when plotting something qucikly, okay to use the pypolot method
* when plotting something more advacned, use the oo method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
# oo method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
           x='age',
           y='chol',
           c='target',
             ax=ax
           );
#ax.set_xlim([45,100]);
# evn the above implementation is not exactly the oo method, it is still the pyplot method
# we have kind of combined the two
# lets do the oo method from scratch
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol");

# add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
# subplots of chol, age, thalach
over_50.head()

fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["thal"])
ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol");

ax.legend(*scatter.legend_elements(), title="Target");

ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True);

scatter1 = ax1.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
ax1.set(title="Heart Disease and Cholestrol Levels",
      #xlabel="Age",
      ylabel="Cholestrol");

# add a legend
ax1.legend(*scatter1.legend_elements(), title="Target");

ax1.axhline(over_50["chol"].mean(),
          linestyle='--');

scatter2 = ax2.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])
ax2.set(title="Heart Disease and Thalach Levels",
      xlabel="Age",
      ylabel="Thalach");

# add a legend
ax2.legend(*scatter2.legend_elements(), title="Target");

ax2.axhline(over_50["thalach"].mean(),
          linestyle='--');

# add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
